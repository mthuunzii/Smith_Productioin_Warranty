# Smiths Manufacturing - Production Warranty Correlation Dashboard

## 📋 Project Overview

This interactive dashboard enables manufacturing engineers to identify root causes of warranty claims by correlating production conditions with defect data. The solution integrates two separate SQLite databases (production and warranty) into a unified analytical view.

**Key Features:**
- Real-time correlation of production parameters with warranty claims
- Statistical trend analysis (OLS regression) to prove temperature-defect relationships
- Interactive visualisations for root cause identification
- Drill-down capability to batch-level data
